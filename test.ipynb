```python
from thinkdsp import decorate,SinSignal
sin_sig=SinSignal(freq=440,amp=1.0,offset=0)
Wave=sin_sig.make_wave(duration=1.0,start=0,framerate=44100)
Wave.plot()
decorate(xlabel="sec",ylabel="amp",title="Wave")
```

```python
segment=Wave.segment(0,1)
spectrum=segment.make_spectrum()
spectrum.plot(high=1000)
decorate(xlabel="freq(Hz)")
```
