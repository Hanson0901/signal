```python
from thinkdsp import CosSignal,SinSignal,decorate
cos_sig=CosSignal(freq=1000,amp=2,offset=0)
sin_sig=SinSignal(freq=800,amp=1.5,offset=0)
```

```python
cos_sig.plot()
decorate(xlabel='Time (s)')
```

```python
sin_sig.plot()
decorate(xlabel='Time (s)')
```

```python
mix = sin_sig + cos_sig
mix
```

```python
mix.plot()
decorate(xlabel='Time (s)')
```
