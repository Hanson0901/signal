```python
import sys
sys.path.append("..")
from thinkdsp import read_wave
filename="./src/lxu2_93911.wav"
```

```python
def stretch(wave, factor):
    wave.ts *= factor
    wave.framerate /= factor
```

```python
wave=read_wave(filename)
stretch(wave, 0.5)
wave.make_audio()
```
